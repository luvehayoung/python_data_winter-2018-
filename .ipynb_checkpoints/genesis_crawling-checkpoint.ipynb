# 제네시스 안전관련 평판 조사
## 보배드림에서 안전관련 게시글 중 제네시스, bmw, 벤츠에 관한 긍정 부정 인식 조사

```python
from bs4 import BeautifulSoup
```

```python
from urllib.request import urlopen
import numpy as np
import pandas as pd
```

```python
#page 10 까지 글
key_url = 'http://www.bobaedream.co.kr/list?code=freeb&s_cate=&maker_no=&model_no=&or_gu=10&or_se=desc&s_selday=&pagescale=30&info3=&noticeShow=&s_select=Subject&s_key=%EC%95%88%EC%A0%84&level_no=&vdate=&type=list&page='
soup = [] #soup객체 넣는 리스트
page_count = 21
for i in range(page_count):
    url = key_url + str(i+1)
    html = urlopen(url)
    soup.append(BeautifulSoup(html,'lxml'))       
```

```python
titles = []
hrefs = []
href = []
for i in range(page_count):
    titles.append(soup[i].select(
        '.bsubject'
    ))

    
for i in titles:
    href = []
    for j in i:
        href.append(j.get('href'))
    hrefs.append(href)
print(hrefs)
#hrefs = hrefs[7:]


```

```python
#href공지사항 빼고
hrefs = np.array(hrefs)
len(hrefs)
```

```python
print(hrefs)
```

```python
fixed = 'http://www.bobaedream.co.kr'
```

```python
ful_url = []
#중간에 하나씩 빼기

for i in range(len(hrefs)):
    if i == 0:
        for j in hrefs[i][8:]:
            ful_url.append(fixed+j)
    else:
        for j in hrefs[i][1:]:
            ful_url.append(fixed+j)
ful_url
        
```

```python
each_titles = []
each_contents = []

for each_url in ful_url: 
    html2 = urlopen(each_url)
    soup2 = BeautifulSoup(html2,'lxml')
    
    tmp = soup2.select(
        '.writerProfile strong'
    )
    
    if len(tmp) != 0:
        each_titles.append(tmp[0].get_text())
        each_contents.append(soup2.select(
            '.bodyCont'
        )[0].get_text())
  
```

```python
print(each_titles)
print(each_contents)
```

```python
result = {
    "제목" : each_titles,
    "내용" : each_contents
}
columns = ["제목","내용"]
df = pd.DataFrame(result, columns=columns)
df
```

```python
df.to_csv('제네시스_안전.csv')
```
